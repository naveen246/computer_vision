```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

%matplotlib inline
```

```python
def read_img(img, read_mode = cv2.IMREAD_COLOR):
    return cv2.imread('input/' + img, read_mode)

def write_img(filename, img):
    cv2.imwrite('output/' + filename, img)
    
def imshow(img):
    plt.axis("off")
    if len(img.shape) > 2:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    else:
        plt.imshow(img)
        
def show_image(img):
    imshow(img)
    plt.show()
    
def show_images_together(img1, img2):
    plt.subplot(121), imshow(img1)
    plt.subplot(122), imshow(img2)
    plt.show()
```

```python
# 1. Input images

img1 = read_img('img1.png')
img2 = read_img('img2.png')

write_img('ps0-1-a-1.png', img1)
write_img('ps0-1-a-2.png', img2)
```

```python
# 2. Color  planes
# a. Swap the red and blue pixels

def swap_red_blue(img):
    img_new = img.copy()
    img_new[:,:,0] = img[:,:,2]
    img_new[:,:,2] = img[:,:,0]
    return img_new

img1_new = swap_red_blue(img1)
write_img('ps0-2-a-1.png', img1_new)

show_images_together(img1, img1_new)
```

```python
# 2 b,c. Create monochrome image

def monochrome(img, color):
    color_index = {'b': 0, 'g': 1, 'r': 2}
    return img[:,:,color_index[color]].copy()

img1_green = monochrome(img1, 'g')
img1_red = monochrome(img1, 'r')
img1_blue = monochrome(img1, 'b')

write_img('ps0-2-b-1.png', img1_green)
write_img('ps0-2-c-1.png', img1_red)

show_images_together(img1_green, img1_red)
```

```python
# 3. Replacement of pixels

def get_center(img, height, width):
    h, w = img.shape
    top = int((h - height) / 2)
    left = int((w - width) / 2)
    return (top, top + height, left, left + width)

def replace_center_pixels(src_img, target_img, center_h, center_w):
    target = target_img.copy()
    top1, bottom1, left1, right1 = get_center(src_img, center_h, center_w)
    top2, bottom2, left2, right2 = get_center(target_img, center_h, center_w)
    target[top2:bottom2, left2:right2] = src_img[top1:bottom1, left1:right1]
    return target

img2_green = monochrome(img2, 'g')
target_center_replaced = replace_center_pixels(img1_green, img2_green, 100, 100)
write_img('ps0-3-a-1.png', target_center_replaced)

show_images_together(target_img, target_center_replaced)
```

```python
# 4. Arithmetic and Geometric operations
# 4a
print('max = ', img1_green.max())
print('min = ', img1_green.min())
print('mean = ', img1_green.mean())
print('std = ', img1_green.std())
```

```python
# 4b
mean = img1_green.mean()
std = img1_green.std()
img1_new = ((img1_green - mean) * 10 / std) + mean
write_img('ps0-4-b-1.png', img1_new)
show_images_together(img1_green, img1_new)
```

```python
# 4 c,d Shift to left and subtract

def shift_img(img, shift_x, shift_y):
    M = np.float32([[1, 0, shift_x], [0, 1, shift_y]])
    img_new = cv2.warpAffine(img, M, img.shape[::-1])
    return img_new

img1_shift = shift_img(img1_green, -2, 0)
img1_sub = cv2.subtract(img1_green, img1_shift, img1_sub)
write_img('ps0-4-c-1.png', img1_shift)
write_img('ps0-4-d-1.png', img1_sub)

show_images_together(img1_shift, img1_sub)
```

```python
# 5 Noise

def add_noise_to_channel(img, color = 'g'):
    noise_sigma = 1
    noise = np.random.normal(scale = noise_sigma, size = img[:,:,0].shape).astype('uint8')
    color_index = {'b': 0, 'g': 1, 'r': 2}
    img_new = img.copy()
    img_new[:,:,color_index[color]] += noise
    return img_new

img1_green_noisy = add_noise_to_channel(img1, 'g')
img1_blue_noisy = add_noise_to_channel(img1, 'b')
write_img('ps0-5-a-1.png', img1_green_noisy)
write_img('ps0-5-b-1.png', img1_blue_noisy)

show_images_together(img1_green_noisy, img1_blue_noisy)
```
